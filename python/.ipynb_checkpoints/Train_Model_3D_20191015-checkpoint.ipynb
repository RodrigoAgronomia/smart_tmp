```python
import os
import cv2
import time
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as TT
import matplotlib.pyplot as plt
from torch.autograd import Variable
from livelossplot import PlotLosses
from torchsummary import summary
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
from cnn.Model import EESPNet_Reg, PredModel
```

```python
class MyDataset(torch.utils.data.Dataset):
    '''
    Class to load the dataset
    '''
    def __init__(self, im_dir, im_list, lbl_list, ps_list, transform = None, img_transform=None):
        '''
        :param imList: image list (Note that these lists have been processed and pickled using the loadData.py)
        :param labelList: label list (Note that these lists have been processed and pickled using the loadData.py)
        '''          
        self.im_dir = im_dir
        self.im_list = im_list
        self.lbl_list = lbl_list
        self.ps_list = ps_list
        self.transform = transform
        self.img_transform = img_transform
        

    def __len__(self):
        return len(self.im_list)


    
    def _calc_pad(self, shp):
        shpd = np.array([400, 200]) - shp
        shpl = shpd - np.int0(shpd/2)
        shpr = shpd - shpl
        shpp = ((0,0), (shpl[0], shpr[0]), (shpl[1], shpr[1]))
        return(shpp)

    def __transform__(self, x):
        x = x.copy()
        x = cv2.resize(x, (np.min([200, int(x.shape[1]/4)]), np.min([400, int(x.shape[0]/4)])))
        x = np.moveaxis(x, 2, 0)
        x = np.pad(x, self._calc_pad(np.array(x.shape[1:])), 'constant', constant_values=(255))
        x = x.astype('float')
        x -= 128
        x /= 35
        return(x)

    def __back_transform__(self, x):
        x = x.copy()
        x *= 35
        x += 128
        x = np.clip(x, 0, 255)
        x = x.astype('uint8')
        x = np.moveaxis(x, 0, 2)
        return(x)
      
    def __getitem__(self, idx):
        '''

        :param idx: Index of the image file
        :return: returns the image and corresponding label file.
        '''
        image_name = self.im_list[idx]
        label = self.lbl_list[idx]
        
        image = cv2.imread(os.path.join(self.im_dir, image_name), cv2.IMREAD_COLOR)
        
        if self.transform:
            image = self.transform(image)

        if self.img_transform:
            image = self.img_transform(image)
            
        
        image = self.__transform__(image)
        image = torch.from_numpy(image)
        label = torch.from_numpy(label)

        image = Variable(image).to(device, dtype=torch.float)
        labelf = Variable(label[:-1]).to(device, dtype=torch.float)
        labelc = Variable(label[-1]).to(device, dtype=torch.long)
        
        return (image, labelf, labelc, self.ps_list[idx])
```

```python
img_transform = TT.Compose([
    TT.ToPILImage(),
    TT.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.05, hue=0.05),
    TT.ToTensor()
])
```

```python
# Diretório das imagens
data_dir  = '../../Datasets/Eucalyptus3/15_10_2019'
im_dir  = '../../Datasets/Eucalyptus3/15_10_2019/frames/'
# Lê o diâmetro das mudas no arquivo csv
plant_params = pd.read_csv(os.path.join(data_dir, 'parametros_medidos.csv'), sep = ',', decimal= '.', header = 0)
```

```python
plant_params
```

```python
im_files = [[os.path.join(p, f), int(p), f] for p in os.listdir(im_dir) for f in os.listdir(os.path.join(im_dir, p))]
imdf = pd.DataFrame(im_files, columns=['img', 'planta', 'frame'])
imfs = pd.merge(imdf, plant_params, how = 'left', on = 'planta')

# Cria os dataframes e cria os conjuntos de treinamento e validação
plant_val = np.random.choice(imfs.planta, 10, replace=False)
im_val = np.in1d(imfs.planta, plant_val)
im_train = np.invert(im_val)
im_train[:] = True
```

```python
train_dataset =  MyDataset(im_dir, imfs.values[im_train,0], imfs.values[im_train, 3:].astype('float'), imfs.values[im_train, 1])

val_dataset =  MyDataset(im_dir, imfs.values[im_val,0], imfs.values[im_val, 3:].astype('float'), imfs.values[im_val, 1])
```

```python
print(len(train_dataset), len(val_dataset))
```

```python
# Carrega o dataset de treinamento
train_loader = torch.utils.data.DataLoader(train_dataset,
                          batch_size=100,
                          shuffle=True,
                          num_workers=0
                         )

# Carrega o dataset de validação
val_loader = torch.utils.data.DataLoader(val_dataset,
                          batch_size=100,
                          shuffle=True,
                          num_workers=0
                         )
```

```python
fct_dim = [16, 8, 4]
fct_dropout = 0.1
fct_dim = [[x, fct_dropout] for x in fct_dim]

n_out = [5, 4]
# Atribui o modelo
modelp = PredModel(fct_dim, n_out)
modelp = modelp.to(device)
modelp.load_state_dict(torch.load('../data/PredModel_tmpp.pth',map_location=device), strict=False)

summary(modelp, input_size= (16,5))
```

```python
# Atribui o modelo
model = EESPNet_Reg(n_out = 5, s = 0.5)
model = model.to(device)
model.load_state_dict(torch.load('../data/FeatModel_tmpp.pth' ,map_location=device), strict=False)
summary(model, input_size=(3, 400, 200))
```

```python
learning_rate = 1e-3
criterion = torch.nn.MSELoss()
criterionc = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(list(model.parameters()) + list(modelp.parameters()), learning_rate)
```

```python
min_loss = 1e9
bad_epochs = 0

liveloss = PlotLosses()
for epoch in range(100):
    bad_epochs += 1
    start_time = time.time()
    epoch_loss = []
    val_loss = []
    
    model.train()
    modelp.train()
    for (imgs, labels, labelsc, ps) in train_loader:
        
        #-------------------------------------------
        # Obtem o output
        pred, feats = model(imgs)
        loss = criterion(pred, labels)
        
        n = len(np.unique(ps))
        for p in np.unique(ps):
            featc = torch.unsqueeze(feats[ps == p].transpose(0,1), 0)
            pad = np.max([0, 5 - featc.shape[2]])
            featc = F.pad(featc, (0, pad), 'constant', 0)
            predf, predfc = modelp(featc)
            loss += criterion(predf, labels[ps == p][[0]])/n
            loss += criterionc(predfc, labelsc[ps == p][[0]])/n

        # Limpa os gradientes
        optimizer.zero_grad()

        # Obtem os gradientes
        loss.backward()

        # Atualiza os parâmetros
        optimizer.step()

        epoch_loss.append(loss.data.item())
    
    model.eval()
    modelp.eval()
    with torch.no_grad():
        for (imgs, labels, labelsc, ps) in val_loader:
            pred, feats = model(imgs)
            loss = criterion(pred, labels)
            
            n = len(np.unique(ps))
            for p in np.unique(ps):
                featc = torch.unsqueeze(feats[ps == p].transpose(0,1), 0)
                pad = np.max([0, 5 - featc.shape[2]])
                featc = F.pad(featc, (0, pad), 'constant', 0)
                predf, predfc = modelp(featc)
                loss += criterion(predf, labels[ps == p][[0]])/n
                loss += criterionc(predfc, labelsc[ps == p][[0]])/n

            val_loss.append(loss.data.item())
                
    epoch_loss = np.array(epoch_loss).mean()
    val_loss = np.array(val_loss).mean()
    time_taken = time.time() - start_time
    
    #save the model
    model_file_name = '../data/FeatModel_tmpp.pth'
    modelp_file_name = '../data/PredModel_tmpp.pth'
    if val_loss < min_loss:
        torch.save(model.state_dict(), model_file_name)
        torch.save(modelp.state_dict(), modelp_file_name)
        min_loss = val_loss
        bad_epochs = 0
    if bad_epochs > 100:
        break

    liveloss.update({
        'log loss': epoch_loss,
        'val_log loss': val_loss
    })
    liveloss.draw()
    print('[%d:%d] loss: %.3f time:%.2f' % (bad_epochs, epoch, epoch_loss, time_taken))
    
```

```python
# Carrega os pesos de um modelo salvo
model.load_state_dict(torch.load(model_file_name,map_location=device))
modelp.load_state_dict(torch.load(modelp_file_name,map_location=device))

print('Modelo OK')
```

```python
start_time = time.time()

model.eval()

obs = []
preds = []
obsf = []
obsfc = []
obsc = []
predsf = []
predsc = []
imgsl = []

for i, (imgs, labels, labelsc, ps) in enumerate(val_loader):
    with torch.no_grad():
        pred, feats = model(imgs)
        for p in np.unique(ps):
            featc = torch.unsqueeze(feats[ps == p].transpose(0,1), 0)
            pad = np.max([0, 5 - featc.shape[2]])
            featc = F.pad(featc, (0, pad), 'constant', 0)
            predf, predc = modelp(featc)
            predsf.append(predf.cpu().data.numpy())
            predsc.append(predc.argmax(1).cpu().data.numpy())
            obsf.append(labels[ps == p][[0]].cpu().data.numpy())
            obsfc.append(labelsc[ps == p][[0]].cpu().data.numpy())
            imgsl.append(imgs[ps == p][[0]].cpu().data.numpy())

    preds.append(pred.cpu().data.numpy())
    obs.append(labels.cpu().data.numpy())
    obsc.append(labelsc.cpu().data.numpy())

imgsl = np.concatenate(imgsl)
y_obs = np.concatenate(obs)
y_obsc = np.concatenate(obsc)
y_pred = np.concatenate(preds)

y_obsf = np.concatenate(obsf)
y_obsfc = np.concatenate(obsfc)
y_predf = np.concatenate(predsf)
y_predc = np.concatenate(predsc)

time_taken = time.time() - start_time
print('time:%.2f' % (time_taken))
```

```python
n_fig = 6
idx = np.random.choice(np.arange(len(imgsl)), n_fig, replace = False)

f, ax = plt.subplots(nrows=1, ncols=n_fig,figsize=(20,5))
for i in range(n_fig):
    im = val_dataset.__back_transform__(imgsl[idx[i]])
    ax[i].imshow(im[:,:,[2,1,0]])
    ax[i].set_title("Obs: {} - Pred: {}".format(y_obsfc[idx[i]], y_predc[idx[i]]), y=1.05, fontsize=12)
```
